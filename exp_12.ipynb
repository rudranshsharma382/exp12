```python
'''1. NOMINAL DATA

categories without any order

example: Gender,Blood group,Department

2.ORDINAL DATA

Ordinal data is a form of categorical data that has a meaningful order among its categories. But, it lacks any numerical values or a fixed interval that can separate them from each other.
'''
```

```python
#Second dataset
import pandas as pd
df=pd.read_csv("/content/Expt11.csv")
df
```

```python
#1 Count students in each grade
#Find how many students obtained each grade.

print(df['Grade'].value_counts())
```

```python
#2 Count students of each gender

print(df['Gender'].value_counts())
```

```python
#3 count students in each department
print(df['Department'].value_counts())
```

```python
#4 percentage distribution of grades

print(df['Grade'].value_counts(normalize=True)*100)
```

```python
#5 Gender vs grade analysis

print(pd.crosstab(df['Gender'],df['Grade']))
```

```python
#6 department vs gender

print(pd.crosstab(df['Department'],df['Gender']))
```

```python
#7 department vs grade

print(pd.crosstab(df['Department'],df['Grade']))
```

```python
#8 department vs grade percentage

print(pd.crosstab(df['Department'],df['Grade'],normalize='index')*100)


```

```python
#9 Grouping data

print(df.groupby('Department')['Grade'].value_counts())
```

```python
#1 FREQUENCY COUNT
#count how many orders in each category.

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}

df_activity2 = pd.DataFrame(data)
print(df_activity2)
```

```python
#2 PERCENTAGE DISTRIBUTION

print(df_activity2['Category'].value_counts(normalize=True)*100)
```

```python
#3 PAYMENT METHOD

print(df_activity2['Payment_Method'].value_counts())
```

```python
#4 UNIQUE VALUES
#DISPLAY ALL UNIQUE CATGORIES

print(df_activity2['Category'].unique())
```

```python
#5 NUMBER OF UNIQUE VALUES

print(df_activity2['Category'].nunique())
```

```python
#6 CROSS TABULATION

print(pd.crosstab(df_activity2['Category'],df_activity2['Payment_Method']))
```

```python
#7 FILTERING TABLE

print(df_activity2[df_activity2['Category']=='Electronics'])
```

```python
#8 Sorting data

sorted_df = df_activity2.sort_values(by='Category')
print(sorted_df)
```

```python
#9 GROUPING

print(df_activity2.groupby('Category')['Payment_Method'].value_counts())
```
